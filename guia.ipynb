```python
import PIL
from PIL import Image, ImageEnhance

# Leer la imagen y convertirla a RGB
imagen = Image.open("readonly/msi_recruitment.gif").convert('RGB')

# Definir la lista de intensidades a utilizar
intensidades = [0.1, 0.5, 0.9]

# Definir la lista de canales de color a variar
canales_color = ["rojo", "verde", "azul"]


# Crear una lista de imágenes para cada canal de color
imagenes_canal_color = {}
for color in canales_color:
    imagenes_canal_color[color] = []

# Iterar sobre los canales de color e intensidades y crear imágenes para cada combinación
for color in canales_color:
    for intensidad in intensidades:
        imagen_canal = imagen.copy()
        pixeles_canal = imagen_canal.load()
        for i in range(imagen_canal.size[0]):
            for j in range(imagen_canal.size[1]):
                r, g, b = pixeles_canal[i, j]
                if color == "rojo":
                    r = int(r * intensidad)
                elif color == "verde":
                    g = int(g * intensidad)
                elif color == "azul":
                    b = int(b * intensidad)
                pixeles_canal[i, j] = (r, g, b)
        imagenes_canal_color[color].append(imagen_canal)

# Crear una hoja de contacto a partir de diferentes imágenes para cada canal de color
primera_imagen = imagenes_canal_color["rojo"][0]
hoja_contacto = PIL.Image.new(primera_imagen.mode, (primera_imagen.width * 3, primera_imagen.height * 3))
x, y = 0, 0

for color in canales_color:
    for imagen in imagenes_canal_color[color]:
        hoja_contacto.paste(imagen, (x, y) )
        if x + primera_imagen.width == hoja_contacto.width:
            x, y = 0, y + primera_imagen.height
        else:
            x += primera_imagen.width

# Redimensionar y mostrar la hoja de contacto
hoja_contacto = hoja_contacto.resize((int(hoja_contacto.width / 2), int(hoja_contacto.height / 2)))
hoja_contacto.show()
```
