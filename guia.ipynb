## Guia de archivos y directorios

    core> Archivos de configuración  
            > config.py  # Archivo de configuración de la aplicación
            > hashing.py # Archivo que contiene la función para encriptar las contraseñas y verificarlas
            > security.py # Archivo que contiene la función para generar el token de seguridad de la aplicación


    apis> Configuración de la api como sus rutas y generales de la app
        > base.py # Archivo que contiene y hace las rutas de la api

    apis> general_pages>  ###!! Quedará en desuso se cambiara a la carpeta api> version1> route_general_pages.py
                          ##!! > route_homepage.py # Rutas de la página principal 

              version1> route_general_pages.py # Rutas de la página principal 
              version1> route_users.py # Rutas de los usuarios en este caso método POST
              version1> route_jobs.py # Rutas de los trabajos en este caso método POST 
              version1> route_login.py # Rutas de los login en este caso método POST     

    templates> Plantillas html de la aplicación
             > components> 
                        Navbar.html # Componentes para la aplicación
             > general_pages> ***!!! Más adelante se copiará este directorio a api> version1> route_general_pages.py
                            # Plantillas para las páginas generales de la aplicación
                            > homepage.html # Plantilla de la página principal
             > shared> 
                      # Archivo base  de la aplicación
                      base.html   # Archivo base de la aplicación todos heredan de aquí

         
     static> 
          # Guarda los archivos estáticos de la aplicación
          images> 
               # Guarda las imágenes de la aplicación
               > logo.png # Imagen del logo de la aplicación

    db>
      # Contiene los archivos de configuración de la base de datos
           > session.py    # Archivo de configuración de la base de datos para crear la sesiones
           > base_class.py #Clase "abuelo" de la que heredan las demás clases de la base de datos
           > base.py       # Archivo que importa los modelos jobs.py y users.py a demás de la clase abuelo base_class.py
      
           models> 
                 #Contiene los modelos de las tablas de la base de datos
                 > job.py # Modelo de la tabla job
                 > user.py # Modelo de la tabla user
           repositories>
                       # Contiene los repositorios de las tablas de la base de datos el código que crea registros en la base de datos
                       > user.py # Crea los usuarios en la base de datos.
                       > job.py # Crea los trabajos en la base de datos.
                       > login.py # Verifica si existe usuario y si si lo devuelve   

    schemas> Contiene los esquemas para validar los datos de las tablas de la base de datos pydantic
            > jobs.py # Esquema validador de la tabla job
            > users.py # Esquema validador de la tabla user
            > tokens.py # Esquema validador para el token de seguridad de la aplicación
